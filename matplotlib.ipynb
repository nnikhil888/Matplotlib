```python
# outler / outline in box plot

x = [10,20,30,40,50,60,70,120]
plt.boxplot(x)
plt.show()
```

```python
# whis : to join whisker to the max 

x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,whis=2)
plt.show()
```

```python
# change the color of 'o' shape 

x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=["python"],patch_artist=True,sym="g")
plt.show()
```

```python
# shape 'o' to '+'
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=["python"],patch_artist=True,sym="g+")
plt.show()
```

```python
# to hide shape 'o'
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=["python"],patch_artist=True,sym="")
plt.show()
```

```python
# box property change

x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=["python"],patch_artist=True,sym="g",boxprops=dict(color="r"))
plt.show()
```

```python
# 1.cap (min & max line)
# 2. whisker color
# filler

x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=["python"],patch_artist=True,sym="g",capprops=dict(color="r"),whiskerprops=dict(color="r"),
            flierprops=dict(markeredgecolor="r"))
plt.show()
```

```python
# multiple box plot

x = [10,20,30,40,50,60,70,120]
x1 = [10,20,40,50,30,60,90]
plt.boxplot(x,labels=["python"],patch_artist=True,sym="g",capprops=dict(color="r"),whiskerprops=dict(color="r"),
            flierprops=dict(markeredgecolor="r"))
plt.boxplot(x1,labels=["python"],patch_artist=True,sym="g",capprops=dict(color="r"),whiskerprops=dict(color="r"),
            flierprops=dict(markeredgecolor="r"))
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120]
x1 = [10,20,40,50,30,60,90]
y = [x,x1]
plt.boxplot(y,sym="g",capprops=dict(color="r"),whiskerprops=dict(color="r"),
            flierprops=dict(markeredgecolor="r"))

plt.show()
```

```python
# color

x = [1,2,3,4,5]
area = [1,2,3,4,5]
area1 = [2,4,3,4,5]
area2 = [3,4,1,6,5]
l =["area","area1","area2"]

plt.stackplot(x,area,area1,area2,labels=l,colors=["r","g","m"])
plt.legend(loc=2)
plt.show()
```

```python
# baseline

x = [1,2,3,4,5]
area = [1,2,3,4,5]
area1 = [2,4,3,4,5]
area2 = [3,4,1,6,5]
l =["area","area1","area2"]

plt.stackplot(x,area,area1,area2,labels=l,colors=["r","g","m"],baseline="sym")
plt.legend(loc=2)
plt.show()
```

```python
x = [1,2,3,4,5]
area = [1,2,3,4,5]
area1 = [2,4,3,4,5]
area2 = [3,4,1,6,5]
l =["area","area1","area2"]

plt.stackplot(x,area,area1,area2,labels=l,colors=["r","g","m"],baseline="wiggle")
plt.legend(loc=2)
plt.show()
```

```python
x = [1,2,3,4,5]
area = [1,2,3,4,5]
area1 = [2,4,3,4,5]
area2 = [3,4,1,6,5]
l =["area","area1","area2"]

plt.stackplot(x,area,area1,area2,labels=l,colors=["r","g","m"],baseline="zero")
plt.legend(loc=2)
plt.show()
```

```python
# title , xlabel , ylabel

x = [1,2,3,4,5]
area = [1,2,3,4,5]
area1 = [2,4,3,4,5]
area2 = [3,4,1,6,5]
l =["area","area1","area2"]

plt.stackplot(x,area,area1,area2,labels=l,colors=["r","g","m"])
plt.legend(loc=2)
plt.title("python")
plt.grid()
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.show()
```

# Fill_between plot

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.title("python")
plt.xlabel("x_axis")
plt.ylabel("y_axis")

plt.plot(x,y,color="r",label="python")
plt.legend(loc=2)
plt.fill_between(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.title("python")
plt.xlabel("x_axis")
plt.ylabel("y_axis")

plt.plot(x,y,color="r",label="python")
plt.legend(loc=2)
plt.fill_between(x=[2,4],y1=2,y2=4)
plt.show()
```

```python
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])

plt.title("python")
plt.xlabel("x_axis")
plt.ylabel("y_axis")

plt.plot(x,y,color="r",label="python")
plt.legend(loc=2)
plt.fill_between(x,y,color="g",where=(x>=2)&(x<=4),alpha=0.5)
plt.show()
```
